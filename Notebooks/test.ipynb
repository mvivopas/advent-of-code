```python
# dir_list_aux = ['/']
dir_list = ['/']
dir_dict = {}
deep_levels = [0]

i_aux = 0
dir_deep = 0

for element in input[1:]:

    if '$ cd' in element:
        if '..' not in element:
            dir_path = dir_list[deep_levels[dir_deep]] + element[5:] + '/'
            dir_list.append(dir_path)
            dir_deep += 1

            if dir_deep >= len(deep_levels):
                deep_levels.append(len(dir_list)-1)
            else:
                deep_levels[dir_deep] = len(dir_list)-1
            i_aux = 0

        if '..' in element and i_aux <= dir_deep: 
            dir_deep -=1
            i_aux +=1   

    if '$' not in element and any(a.isdigit() for a in element):
        
        if dir_dict.get(dir_path) is not None:
            dir_dict[dir_path].append([int(a) for a in element.split() if a.isdigit()][0])
        else: 
            dir_dict[dir_path] = [int(a) for a in element.split() if a.isdigit()]
    
```

```python
sum_cum = 0
sum_cum_gr = 0

for el in dir_list:
    for key in dir_dict:

        if el in key:
            sum_cum += sum(dir_dict[key])
    
    if sum_cum < 100000:
        sum_cum_gr += sum_cum

    sum_cum = 0  

print(sum_cum_gr)
```
