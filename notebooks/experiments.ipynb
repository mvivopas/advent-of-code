```python
with open('/Users/mariavivo/Desktop/advent_of_code/data/input_p4.txt') as f:
    input = f.readlines()

input = [i.rstrip().split(',') for i in input]


print(input[:10],)
```

```python
start = time.time()
score_cum = 0
for pair1, pair2 in input:
    pair1 = [ int(p) for p in pair1.split('-')]
    pair2 = [ int(p) for p in pair2.split('-')]

    farter_start = max(pair1[0],pair2[0])
    closer_end = min(pair1[1],pair2[1]) 

    if any([farter_start,closer_end] == x for x in [pair1, pair2]):
        score_cum += 1

elapsed_time = time.time() - start 

print(f"Elapsed time: {elapsed_time}")
print(f"Number of totally overlapped pairs is {score_cum}")
```

```python
# best sol
irange = lambda a, b: range(a, b + 1)


def p1(f):
    ans = 0
    for a, b in [input[0]]:
        a = set(irange(*map(int, a.split("-"))))
        b = set(irange(*map(int, b.split("-"))))
        if len(a - b) == 0 or len(b - a) == 0:
            ans += 1
    return ans
```
